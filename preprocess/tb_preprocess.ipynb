```python
%matplotlib inline
import os
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python

# process MontogomerySet
path = "D:\\GoogleDrive\\dataset\\radiology\\tuberculosis_xray\\MontgomerySet\\CXR_png\\"
path_save = "D:\\GoogleDrive\\dataset\\radiology\\tuberculosis_xray\\dataset\\"
size = 512
for filename in os.listdir(path):
    if filename.endswith('.png'):
        filepath = path + filename
        img = cv2.imread(filepath,0)
        img = cv2.resize(img, (size,size))

        img = (img - 255/2)/255*2

        savename = os.path.splitext(filename)[0]
        with open(path_save + savename +'.npy' , 'wb') as f:
            np.save(f, img)

# process ChinaSet_AllFiles
path = "D:\\GoogleDrive\\dataset\\radiology\\tuberculosis_xray\\ChinaSet_AllFiles\\CXR_png\\"
for filename in os.listdir(path):
    if filename.endswith('.png'):
        filepath = path + filename
        img = cv2.imread(filepath,0)
        img = cv2.resize(img, (size,size))

        img = (img - 255/2)/255*2

        savename = os.path.splitext(filename)[0]
        with open(path_save + savename +'.npy' , 'wb') as f:
            np.save(f, img)
```

```python
# Create csv file
path = "D:\\GoogleDrive\\dataset\\radiology\\tuberculosis_xray\\"

df = pd.read_csv(path+  'label.csv')
df = df.drop('Unnamed: 0', axis= 1)
df_train, df_test = train_test_split(df, test_size = 0.2, random_state = 42)

df_train.to_csv(path + 'df_train.csv', index = False)
df_test.to_csv(path + 'df_test.csv', index = False)
```
